```python
import csv 
import sys
import pandas as pd

transfer_metrics = pd.read_csv('./files/CA_MetricsforMap_FINAL(Applicant Type_Major).csv')
transfer_metrics = transfer_metrics[['Course ID', 'Dept Cd', 'Breakdown', 'Disproportionate Impact']]
transfer_metrics.drop(transfer_metrics[transfer_metrics['Dept Cd'] == 'No major'].index, inplace = True)
transfer_metrics.drop(transfer_metrics[transfer_metrics['Breakdown'] == 'Unknown'].index, inplace = True)

total_metrics = pd.read_csv('./files/CA_MetricsforMap_FINAL(Metrics).csv')
total_metrics.insert(loc=6, column='Transfer Disproportionate Impact', value='')

transfer_metrics.groupby(['Course ID', 'Dept Cd'])
print(transfer_metrics)
```

```python
for id in transfer_metrics['Course ID'].unique():
    transfer_filtered = transfer_metrics[transfer_metrics['Course ID'] == id]
    for dept in transfer_filtered['Dept Cd'].unique():
        transfer_single = transfer_filtered[transfer_filtered['Dept Cd'] == dept]

        # Case 1: Only one year's data -> skip
        if (len(transfer_single) == 1):
            continue
        else:
            if (transfer_single.iloc[0,3] != 'Y' and transfer_single.iloc[1,3] == 'Y'):
                total_metrics.loc[(total_metrics['Course ID'] == id) & (total_metrics['Dept Cd'] == dept), 'Transfer Disproportionate Impact'] = 'Y'
                print(transfer_single)
                
        # else:
        #     if (transfer_single.iloc[1, 3] == 'Y'):
        #         print(transfer_single)

        # if (transfer_single.iloc[0].Breakdown == '' and transfer_single.iloc[1].Breakdown == 'Y'):
        #     print(transfer_single)
```
